# LeNet-5 Model for MNIST

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Load and Preprocess the MNIST Dataset

We will load the MNIST dataset, which consists of grayscale images of handwritten digits (0-9). The images are 28x28 pixels. We need to normalize the pixel values to the range [0, 1] and reshape the images to include a channel dimension (28x28x1).

```python
# Load MNIST dataset
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

## 2. Define the LeNet-5 Model Architecture

The LeNet-5 architecture is a classic convolutional neural network designed by Yann LeCun. It consists of:

1.  **C1 (Convolutional Layer)**: 6 filters, 5x5 kernel, stride 1, ReLU activation.
2.  **S2 (Average Pooling Layer)**: 2x2 kernel, stride 2.
3.  **C3 (Convolutional Layer)**: 16 filters, 5x5 kernel, stride 1, ReLU activation.
4.  **S4 (Average Pooling Layer)**: 2x2 kernel, stride 2.
5.  **C5 (Convolutional Layer)**: 120 filters, 5x5 kernel, stride 1, ReLU activation (fully connected).
6.  **F6 (Fully Connected Layer)**: 84 units, ReLU activation.
7.  **Output Layer**: 10 units (for 10 classes), Softmax activation.

Note: The original LeNet-5 used tanh activation and specific connection patterns, but we'll use a common modern adaptation with ReLU and standard connections for simplicity.

```python
def create_lenet5_model():
    model = models.Sequential([
        # C1: Convolutional Layer
        layers.Conv2D(filters=6, kernel_size=(5, 5), activation='tanh', input_shape=(28, 28, 1)),
        # S2: Average Pooling Layer
        layers.AveragePooling2D(pool_size=(2, 2), strides=(2, 2)),
        # C3: Convolutional Layer
        layers.Conv2D(filters=16, kernel_size=(5, 5), activation='tanh'),
        # S4: Average Pooling Layer
        layers.AveragePooling2D(pool_size=(2, 2), strides=(2, 2)),
        # Flatten the feature maps for the fully connected layers
        layers.Flatten(),
        # C5: Fully Connected Layer (originally convolutional with 5x5 filters and 1x1 input)
        layers.Dense(units=120, activation='tanh'),
        # F6: Fully Connected Layer
        layers.Dense(units=84, activation='tanh'),
        # Output Layer
        layers.Dense(units=10, activation='softmax')
    ])
    return model

# Create the model
model = create_lenet5_model()

# Display the model summary
model.summary()
```
