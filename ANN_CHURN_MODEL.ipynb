### 1. **Importing Keras**

**Analogy:** Keras is like a set of LEGO instructions. Instead of carving plastic blocks yourself, you just snap pre-made pieces together.

### 2. **The Sequential Model**

```python
from keras.models import Sequential
```

| Part | What It Means |
|------|---------------|
| **keras.models** | A sub-folder inside Keras that contains different "types" of neural networks |
| **Sequential** | The **simplest type** of neural network — data flows in a straight line |

**What is a Sequential Model?**

```
Input → [Layer 1] → [Layer 2] → [Layer 3] → Output
```

- Data enters from the left
- Passes through each layer one by one
- Comes out as a prediction on the right

**Analogy:** An assembly line in a factory. Raw materials go in, get processed step by step, finished product comes out.

**When to use Sequential:**
- ✅ Your data flows in one direction (most common)
- ✅ Each layer connects to exactly one next layer
- ❌ Not for: models with multiple inputs, multiple outputs, or layers that branch/skip

**Other model types (not imported here):**
- `Functional` API — for complex architectures (multiple inputs, shared layers)
- `Model` subclassing — for maximum customization

---

### 3. **The Dense Layer**

```python
from keras.layers import Dense
```

| Part | What It Means |
|------|---------------|
| **keras.layers** | A sub-folder containing different "types" of layers you can add to your network |
| **Dense** | A **fully connected layer** — every neuron connects to every neuron in the next layer |

**What is a Dense Layer?**

```
        Neuron 1
       /   |   \
Input ----Neuron 2---- Output
       \   |   /
        Neuron 3
```

- Every input connects to **every** neuron in the layer
- Every neuron in the layer connects to **every** output
- No skipping, no shortcuts

**Analogy:** A group chat where everyone can message everyone else. VS. A broadcast system where only some people talk.

**Why "Dense"?**
- Because the connections are "densely packed" — maximum connectivity
- Also called "fully connected" or "FC layer"

**Other layer types (not imported here, but common):**

| Layer | What It Does | When to Use |
|-------|------------|-------------|
| **Conv2D** | Looks at small patches of images | Image recognition |
| **LSTM** | Remembers sequences over time | Text, speech, time series |
| **Dropout** | Randomly turns off neurons during training | Prevent overfitting |
| **BatchNormalization** | Standardizes layer inputs | Speed up training |
| **Flatten** | Converts 2D to 1D | Between Conv2D and Dense |

## Common Beginner Questions

**Q: Why not just `import keras`? Why the specific imports?**

- `import keras` loads EVERYTHING (heavy, slower)
- `from keras.models import Sequential` only loads what you need (cleaner, faster)
- Also avoids typing `keras.models.Sequential()` every time — just `Sequential()`

**Q: What's the difference between Keras and TensorFlow?**

| TensorFlow | Keras |
|------------|-------|
| The "engine" — does the heavy math | The "dashboard" — easy interface |
| Created by Google | Originally separate, now built into TF |
| Complex, powerful | Simple, user-friendly |
| You CAN use TF directly | Most people use TF through Keras |

Since TensorFlow 2.0, Keras is the **official recommended way** to use TensorFlow.

**Q: Can I build a neural network without these imports?**

Technically yes, but you'd need to:
- Write thousands of lines of math code
- Handle matrix multiplication manually
- Implement backpropagation from scratch
- Debug complex calculus

Keras saves you from all that!

## `classifier.compile(...)`

### `optimizer='adam'`

| Aspect | Description |
|---|---|
| **What it is** | Adaptive Moment Estimation — combines momentum and RMSProp |
| **What it does** | Adjusts learning rate per-parameter based on gradient history |
| **Default learning rate** | 0.001 |
| **Alternatives** | `'sgd'`, `'rmsprop'`, `'adagrad'`, `'adamw'` |
| **When to change** | Use `'adamw'` if you need built-in weight decay; `'sgd'` with momentum for fine-tuning |

### `loss='binary_crossentropy'`

| Aspect | Description |
|---|---|
| **What it is** | Log loss for binary classification |
| **Formula** | `−[y·log(p) + (1−y)·log(1−p)]` |
| **What it does** | Penalizes confident wrong predictions heavily |
| **Requires** | Labels as 0 or 1; output layer with `sigmoid` activation |
| **Alternatives** | `'categorical_crossentropy'` (one-hot labels), `'sparse_categorical_crossentropy'` (integer labels), `'mse'` (regression) |

### `metrics=['accuracy']`

| Aspect | Description |
|---|---|
| **What it is** | List of metrics to evaluate during training/validation |
| **What it does** | Computes accuracy: `correct predictions / total predictions` |
| **Displayed as** | `accuracy` (training), `val_accuracy` (validation) |
| **Alternatives** | `['precision']`, `['recall']`, `['auc']`, `['mae']` |
| **Note** | Not used for gradient computation — only for monitoring |

````python
#Importing few libraries for k-folds
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import cross_val_score
from keras.models import Sequential
from keras.layers import Dense
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

def build_classifier():
    classifier = Sequential()
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 10))
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
    classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier

keras_model = KerasClassifier(build_fn = build_classifier, batch_size = 10, epochs = 10, verbose=0)

# Create a Pipeline to impute missing values before passing to KerasClassifier
pipeline = Pipeline([
    ('imputer', SimpleImputer(missing_values=float('nan'), strategy='mean')),
    ('keras_model', keras_model)
])

accuracies = cross_val_score(estimator = pipeline, X = X_train, y = y_train, cv =10, n_jobs = -1 )
mean1 = accuracies.mean()
variance = accuracies.std()
````

`GridSearchCV` is a powerful, automated tool in the `scikit-learn` library used for **hyperparameter tuning**. It helps you find the optimal configuration for your machine learning model by systematically exploring a predefined "grid" of hyperparameter values.

### Why do we need it?

In machine learning, your model's performance often depends heavily on its **hyperparameters**—settings you choose *before* training begins (like the learning rate, the number of trees in a forest, or the depth of a decision tree). Since there is no "magic" way to know the perfect values for these in advance, engineers historically had to guess or perform manual testing. `GridSearchCV` automates this search process.

### How it works

`GridSearchCV` combines two core concepts:

1. **Grid Search:** It creates a table (a grid) of all possible combinations of the hyperparameters you provide.
2. **Cross-Validation:** For every single combination in that grid, it performs **K-Fold Cross-Validation** to evaluate the model's performance.

### The Workflow

1. **Define the Grid:** You create a dictionary where keys are the hyperparameter names and values are the lists of settings you want to test.
2. **Initialize the Estimator:** You tell `GridSearchCV` which model you want to tune.
3. **Fit:** When you call `.fit()`, the tool trains and validates the model for every combination in your grid.
4. **Extract Best Results:** Once finished, it identifies the combination that produced the best score (e.g., highest accuracy) and allows you to access it using attributes like `.best_params_`.

### Key Considerations

* **Computational Cost:** Because `GridSearchCV` tries every single combination, it can be extremely slow if your grid is large or your dataset is massive.
* **Parallelization:** You can use the `n_jobs=-1` parameter to tell the tool to use all available CPU cores to speed up the process.
* **The "Grid" Limitation:** It only searches the combinations you explicitly define. If the "perfect" value for a parameter lies somewhere in between two values you specified, `GridSearchCV` won't find it.

### Example in Python

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

# Define your parameter grid
param_grid = {
    'C': [0.1, 1, 10],
    'gamma': [1, 0.1, 0.01]
}

# Initialize the grid search
grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=2, cv=5)

# Fit the model
grid.fit(X_train, y_train)

# See the best parameters found
print(grid.best_params_)

```
