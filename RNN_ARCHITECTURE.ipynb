# Embeddings in RNNs

Let me break this down clearly.

## What is an Embedding?

An **embedding** is a way to convert discrete tokens (words, characters, IDs) into dense numeric vectors that a neural network can process.

```
"cat"  →  [0.2, -0.5, 0.8, 0.1, ...]   # e.g., 128-dimensional vector
"dog"  →  [0.3, -0.4, 0.7, 0.2, ...]   # similar to "cat" → close in space
"bank" →  [-0.6, 0.9, -0.1, 0.4, ...]  # very different
```

## Why RNNs Need Embeddings

RNNs process **sequences step by step**. At each time step, they need a numeric vector as input. Raw words can't be fed directly — they must be converted first.

```
Sentence: ["I", "love", "ML"]

Step 1: "I"    → embed → [0.1, 0.9, ...]  → RNN cell → hidden state h₁
Step 2: "love" → embed → [0.5, 0.2, ...]  → RNN cell → hidden state h₂
Step 3: "ML"   → embed → [0.8, 0.1, ...]  → RNN cell → hidden state h₃
```

## The Full Picture

```
Input tokens
   ↓
[Embedding Layer]   ← learned lookup table (vocab_size × embed_dim)
   ↓
Dense vectors
   ↓
[RNN / LSTM / GRU]  ← processes sequence step by step
   ↓
Hidden states → Output
```

## In Code (PyTorch)

```python
import torch
import torch.nn as nn

class RNN_with_Embedding(nn.Module):
    def __init__(self, vocab_size, embed_dim, hidden_dim, output_dim):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim)  # lookup table
        self.rnn = nn.RNN(embed_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, output_dim)

    def forward(self, x):          # x: (batch, seq_len) — token IDs
        embedded = self.embedding(x)         # → (batch, seq_len, embed_dim)
        output, hidden = self.rnn(embedded)  # → (batch, seq_len, hidden_dim)
        return self.fc(hidden.squeeze(0))    # → (batch, output_dim)
```

## Key Concepts

| Concept | Meaning |
|---|---|
| `vocab_size` | Total unique tokens (e.g., 10,000 words) |
| `embed_dim` | Size of each vector (e.g., 128 or 300) |
| Lookup table | Each token ID maps to one row in a matrix |
| Learned | Embedding weights are updated during training |
| Pretrained | Can use GloVe / Word2Vec instead of learning from scratch |

## Why Not One-Hot Encoding Instead?

| One-Hot | Embedding |
|---|---|
| Sparse (10,000 zeros + one 1) | Dense (e.g., 128 floats) |
| No semantic meaning | Similar words are close in space |
| Fixed, not learned | Learned during training |
| Huge input to RNN | Compact, efficient |

---

**In short:** The embedding layer acts as a *translator* — converting token IDs into meaningful dense vectors before the RNN processes them sequentially. Both the embedding and RNN weights are learned together end-to-end.

## Breaking Down This RNN Model

This is a **Keras Sequential model** with 2 layers. Let me explain each part.

---

### Layer 1: `SimpleRNN(32, input_shape=(50,1), return_sequences=False)`

```
input_shape = (50, 1)
      │        │
      │        └── each time step has 1 feature (e.g., 1 value per step)
      └─────────── sequence length = 50 time steps
```

```
SimpleRNN(32)
     │
     └── 32 hidden units (neurons in the RNN cell)
```

**What happens inside:**

```
Time step 1:  x₁ [1 value]  ──→ ┌─────────┐
Time step 2:  x₂ [1 value]  ──→ │   RNN   │ ──→ h₁, h₂, ... h₅₀
Time step 3:  x₃ [1 value]  ──→ │  Cell   │
   ...                           │ (32 dim)│
Time step 50: x₅₀ [1 value] ──→ └─────────┘
```

**`return_sequences=False`** → only returns the **last hidden state**

```
# return_sequences=False  →  output shape: (batch, 32)   ← only h₅₀
# return_sequences=True   →  output shape: (batch, 50, 32) ← all h₁...h₅₀
```

---

### Layer 2: `Dense(1, activation='sigmoid')`

```
32 values from RNN
       │
  [Dense Layer]   ← fully connected: 32 → 1
       │
  sigmoid(x) = 1 / (1 + e⁻ˣ)
       │
  output: single value between 0 and 1
```

This means the model is doing **binary classification** (e.g., positive/negative sentiment).

---

### Full Data Flow

```
Input
(batch, 50, 1)        50 time steps, 1 feature each
     ↓
SimpleRNN(32)
(batch, 32)           only last hidden state (return_sequences=False)
     ↓
Dense(1, sigmoid)
(batch, 1)            probability → 0 or 1
```

---

### Concrete Example

If you're doing **sentiment analysis on text:**

```
"I love this movie"

Tokenized + padded → [4, 27, 8, 153, 0, 0, ..., 0]  # 50 tokens
                                                  ↑
                                            padded to length 50

Each token → 1 feature value (or passed through embedding first)
     ↓
RNN processes all 50 steps
     ↓
Dense(sigmoid) → 0.91  →  Positive 👍
```

---

### Summary Table

| Part | Value | Meaning |
|---|---|---|
| `input_shape=(50,1)` | 50 steps × 1 feature | sequence of 50 single values |
| `SimpleRNN(32)` | 32 units | 32-dimensional hidden state |
| `return_sequences=False` | only last output | summarizes whole sequence |
| `Dense(1)` | 1 neuron | single output |
| `sigmoid` | output ∈ (0,1) | binary classification |
