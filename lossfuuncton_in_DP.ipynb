# Deep Learning: Complete Loss & Activation Functions Reference

## PART 1: ACTIVATION FUNCTIONS

### 1.1 Sigmoid Family (S-Shaped, Bounded)

| Function | Formula | Range | Derivative | Key Properties |
|----------|---------|-------|------------|----------------|
| **Sigmoid / Logistic** | $\sigma(x) = \frac{1}{1+e^{-x}}$ | (0, 1) | $\sigma(x)(1-\sigma(x))$ | • Vanishing gradients at extremes<br>• Not zero-centered<br>• Used in output layer for binary classification |
| **Hard Sigmoid** | $\max(0, \min(1, 0.2x + 0.5))$ | [0, 1] | Piecewise constant | • Faster computation<br>• Used in quantized/mobile networks |
| **Swish** | $x \cdot \sigma(x)$ | (-∞, ∞) | Complex | • Self-gated<br>• Smooth, non-monotonic<br>• Often outperforms ReLU |

### 1.2 Tanh Family (S-Shaped, Zero-Centered)

| Function | Formula | Range | Derivative | Key Properties |
|----------|---------|-------|------------|----------------|
| **Tanh** | $\frac{e^x - e^{-x}}{e^x + e^{-x}}$ | (-1, 1) | $1 - \tanh^2(x)$ | • Zero-centered (better than sigmoid)<br>• Still suffers vanishing gradients<br>• Good for RNN hidden states |
| **Hard Tanh** | $\max(-1, \min(1, x))$ | [-1, 1] | 1 or 0 | • Computationally cheaper |

### 1.3 ReLU Family (Rectified Linear — Most Popular)

| Function | Formula | Range | Derivative | Key Properties |
|----------|---------|-------|------------|----------------|
| **ReLU** | $\max(0, x)$ | [0, ∞) | 0 if x<<0, 1 if x>0 | • Simple, fast<br>• **Dying ReLU problem** (neurons stuck at 0)<br>• Not zero-centered |
| **Leaky ReLU** | $\max(\alpha x, x)$, α=0.01 | (-∞, ∞) | α if x<<0, 1 if x>0 | • Fixes dying ReLU<br>• Small negative slope |
| **PReLU** | $\max(\alpha x, x)$, α learned | (-∞, ∞) | Learned α | • Parametric, trainable α |
| **ELU (Exponential LU)** | $x$ if x≥0, $\alpha(e^x-1)$ if x<<0 | (-α, ∞) | 1 if x≥0, $\alpha e^x$ if x<<0 | • Smooth negative region<br>• Mean closer to zero |
| **SELU (Scaled ELU)** | $\lambda x$ if x≥0, $\lambda\alpha(e^x-1)$ if x<<0 | (-λα, ∞) | Self-normalizing | • **Self-normalizing property**<<br>• No batch norm needed<br>• Specific initialization required |
| **GELU (Gaussian Error LU)** | $x \cdot \Phi(x)$ | (-∞, ∞) | Approximated | • Smooth, stochastic relaxation<br>• Used in Transformers (BERT, GPT) |
| **SiLU / Swish** | $x \cdot \sigma(x)$ | (-∞, ∞) | Complex | • Same as Swish<br>• Used in modern architectures |
| **Mish** | $x \cdot \tanh(\text{softplus}(x))$ | (-∞, ∞) | Complex | • Smooth, non-monotonic<br>• Strong empirical performance |

### 1.4 Linear / Identity

| Function | Formula | Range | Use Case |
|----------|---------|-------|----------|
| **Linear** | $f(x) = x$ | (-∞, ∞) | Regression output layer |
| **Identity** | Same as linear | (-∞, ∞) | Autoencoder bottleneck, skip connections |

### 1.5 Softmax Family (Probability Distributions)

| Function | Formula | Range | Use Case |
|----------|---------|-------|----------|
| **Softmax** | $\frac{e^{x_i}}{\sum_j e^{x_j}}$ | (0, 1), sum=1 | Multi-class classification output |
| **Log-Softmax** | $\log(\text{softmax}(x_i))$ | (-∞, 0) | Numerically stable with NLLLoss |
| **Sparsemax** | Euclidean projection onto probability simplex | [0, 1], sum=1 | Sparse attention, exact zeros |

### 1.6 Specialized / Niche Activations

| Function | Formula | Use Case |
|----------|---------|----------|
| **Softplus** | $\log(1 + e^x)$ | Smooth approximation of ReLU |
| **Maxout** | $\max(w_1^T x + b_1, w_2^T x + b_2)$ | Learnable piecewise linear |
| **Softsign** | $\frac{x}{1 + |x|}$ | Alternative to tanh |
| **Arctan** | $\arctan(x)$ | (-π/2, π/2), smooth |
| **Bent Identity** | $\frac{\sqrt{x^2+1}-1}{2} + x$ | Non-monotonic, self-stabilizing |

## PART 2: LOSS FUNCTIONS

### 2.1 Classification Losses

#### Binary Classification

| Loss | Formula | Activation Pairing | Use Case |
|------|---------|-------------------|----------|
| **Binary Cross-Entropy (BCE)** | $-\frac{1}{N}\sum [y\log(\hat{y}) + (1-y)\log(1-\hat{y})]$ | Sigmoid | Standard binary classification |
| **BCE with Logits** | BCE applied to raw logits | None (sigmoid inside) | Numerically stable |
| **Focal Loss** | $-\frac{1}{N}\sum \alpha_t(1-\hat{y}_t)^\gamma \cdot \text{BCE}$ | Sigmoid | Imbalanced datasets (down-weights easy examples) |
| **Hinge Loss** | $\max(0, 1 - y\cdot\hat{y})$ | None (or Tanh) | SVM-style binary classification |

#### Multi-Class Classification (Single Label)

| Loss | Formula | Activation Pairing | Use Case |
|------|---------|-------------------|----------|
| **Categorical Cross-Entropy** | $-\sum_i y_i \log(\hat{y}_i)$ | Softmax | One-hot labels |
| **Sparse Categorical Cross-Entropy** | Same as above | Softmax | Integer labels (memory efficient) |
| **Cross-Entropy with Logits** | Applied to raw logits | None (softmax inside) | Numerically stable |
| **Focal Loss (Multi-class)** | $(1-\hat{y}_t)^\gamma \cdot \text{CE}$ | Softmax | Imbalanced multi-class |
| **Hinge Loss (Multi-class)** | $\sum_{j \neq y} \max(0, 1 + \hat{y}_j - \hat{y}_y)$ | None | SVM-style multi-class |

#### Multi-Label Classification

| Loss | Formula | Activation Pairing | Use Case |
|------|---------|-------------------|----------|
| **Binary Cross-Entropy (per label)** | $\sum_{c=1}^{C} \text{BCE}_c$ | Sigmoid (per class) | Multiple independent labels |
| **Asymmetric Loss** | $-\sum [y(1-\hat{y})^\gamma \log(\hat{y}) + (1-y)\hat{y}^\gamma \log(1-\hat{y})]$ | Sigmoid | Positive-negative imbalance in multi-label |

### 2.2 Regression Losses

| Loss | Formula | Activation Pairing | Properties |
|------|---------|-------------------|------------|
| **Mean Squared Error (MSE / L2)** | $\frac{1}{N}\sum (y - \hat{y})^2$ | Linear | • Penalizes large errors heavily<br>• Sensitive to outliers |
| **Mean Absolute Error (MAE / L1)** | $\frac{1}{N}\sum \|y - \hat{y}\|$ | Linear | • Robust to outliers<br>• Non-differentiable at zero |
| **Huber Loss** | $\begin{cases} \frac{1}{2}(y-\hat{y})^2 & \text{if } \|error\| \leq \delta \\ \delta\|y-\hat{y}\| - \frac{1}{2}\delta^2 & \text{otherwise} \end{cases}$ | Linear | • MSE for small errors, MAE for large<br>• Robust yet smooth |
| **Log-Cosh Loss** | $\frac{1}{N}\sum \log(\cosh(\hat{y} - y))$ | Linear | • Smooth approximation of MAE<br>• Less sensitive to outliers |
| **Quantile Loss** | $\begin{cases} \tau(y-\hat{y}) & \text{if } y \geq \hat{y} \\ (1-\tau)(\hat{y}-y) & \text{otherwise} \end{cases}$ | Linear | • Predicts quantiles<br>• Asymmetric |
| **Tweedie Loss** | Compound Poisson-Gamma log-likelihood | Linear / Exp | • Insurance claims, zero-inflated data |

### 2.3 Probabilistic / Distribution Losses

| Loss | Formula | Use Case |
|------|---------|----------|
| **Kullback-Leibler (KL) Divergence** | $D_{KL}(P\|Q) = \sum P(x) \log\frac{P(x)}{Q(x)}$ | VAEs, knowledge distillation, variational inference |
| **Jensen-Shannon Divergence** | $\frac{1}{2}D_{KL}(P\|M) + \frac{1}{2}D_{KL}(Q\|M)$, $M=\frac{P+Q}{2}$ | GANs (symmetric, bounded) |
| **Wasserstein Distance** | $\inf_{\gamma \in \Gamma(P,Q)} \mathbb{E}_{(x,y)\sim\gamma}[\|x-y\|]$ | WGANs, optimal transport |
| **Negative Log-Likelihood (NLL)** | $-\log P(y\|x)$ | General probabilistic modeling |
| **Cramér Distance** | $\int_{-\infty}^{\infty} \|F_P(x) - F_Q(x)\|^2 dx$ | Distributional RL |

### 2.4 Ranking & Metric Learning Losses

| Loss | Formula | Use Case |
|------|---------|----------|
| **Triplet Loss** | $\max(0, d(a,p) - d(a,n) + \margin)$ | Face recognition, similarity learning |
| **Contrastive Loss** | $\frac{1}{2}y \cdot d^2 + \frac{1}{2}(1-y)\max(0, \margin-d)^2$ | Siamese networks |
| **ArcFace / CosFace** | Additive angular margin softmax | Face recognition (state-of-the-art) |
| **InfoNCE / NT-Xent** | $-\log\frac{\exp(\text{sim}(z_i,z_j)/\tau)}{\sum_k \exp(\text{sim}(z_i,z_k)/\tau)}$ | Contrastive learning (SimCLR, CLIP) |

### 2.5 Segmentation & Structured Prediction

| Loss | Formula | Use Case |
|------|---------|----------|
| **Dice Loss** | $1 - \frac{2|X \cap Y|}{|X| + |Y|}$ | Medical image segmentation |
| **IoU Loss / Tversky Loss** | $1 - \frac{|X \cap Y|}{|X \cup Y|}$ | Object detection, segmentation |
| **Focal Loss (Segmentation)** | Asymmetric weighting of CE | Imbalanced segmentation |
| **Lovász-Softmax** | Convex surrogate of IoU | Direct optimization of IoU metric |
| **Boundary Loss** | Distance-based term | Boundary-aware segmentation |

### 2.6 Sequence & Language Model Losses

| Loss | Formula | Use Case |
|------|---------|----------|
| **Cross-Entropy (Token-level)** | $-\sum_t \log P(y_t \| y_{<<t}, x)$ | Language modeling, machine translation |
| **CTC Loss** | $-\log \sum_{\pi \in \mathcal{B}^{-1}(y)} \prod_t p(\pi_t \| x_t)$ | Speech recognition, handwriting |
| **Connectionist Temporal Classification** | Same as above | Alignment-free sequence learning |

### 2.7 Adversarial & Generative Losses

| Loss | Formula | Use Case |
|------|---------|----------|
| **Minimax GAN Loss** | $\min_G \max_D V(D,G)$ | Standard GANs |
| **Non-Saturating GAN Loss** | $-\mathbb{E}[\log(D(G(z)))]$ | Generator training (avoid vanishing) |
| **Wasserstein GAN Loss** | $\mathbb{E}[D(x)] - \mathbb{E}[D(G(z))]$ | Training stability |
| **LSGAN Loss** | $\frac{1}{2}\mathbb{E}[(D(x)-1)^2] + \frac{1}{2}\mathbb{E}[D(G(z))^2]$ | Least squares GAN |
| **Perceptual Loss** | $\|\phi(I) - \phi(\hat{I})\|$ (VGG features) | Style transfer, super-resolution |

### 2.8 Regularization & Auxiliary Losses

| Loss | Formula | Purpose |
|------|---------|---------|
| **L1 Regularization** | $\lambda \sum \|w\|$ | Sparse weights, feature selection |
| **L2 Regularization / Weight Decay** | $\lambda \sum w^2$ | Weight shrinkage, generalization |
| **Dropout** | Random zeroing (implicit ensemble) | Prevent co-adaptation |
| **Label Smoothing** | $-\sum [(1-\epsilon)y + \epsilon/K] \log(\hat{y})$ | Prevent overconfidence |
| **Mixup Loss** | $\lambda \text{Loss}(x_1,y_1) + (1-\lambda)\text{Loss}(x_2,y_2)$ | Data augmentation, robustness |

## PART 3: VALID PAIRING MATRIX

| Task | Output Activation | Loss Function | Notes |
|------|-------------------|---------------|-------|
| **Binary Classification** | Sigmoid | BCE | Standard |
| **Multi-Class (Single Label)** | Softmax | Categorical CE | One-hot or integer labels |
| **Multi-Label** | Sigmoid (per class) | BCE (sum over classes) | Independent probabilities |
| **Regression** | Linear | MSE / MAE / Huber | Unbounded real values |
| **Positive Regression** | ReLU / Softplus | MSE | Non-negative outputs |
| **Bounded Regression [0,1]** | Sigmoid | MSE / BCE | Normalized targets |
| **Probability Distribution** | Softmax | KL Divergence | Distribution matching |
| **Sequence Generation** | Softmax | Token-level CE | Autoregressive models |
| **Metric Learning** | Linear + L2-Norm | Triplet / Contrastive | Embedding space |
| **Image Segmentation** | Sigmoid / Softmax | Dice / IoU / CE | Pixel-wise classification |
| **Generative Modeling** | Sigmoid (discriminator) | Adversarial losses | GAN framework |

## PART 4: Quick Decision Guide

```
Is it classification?
  ├── Binary? → Sigmoid + BCE
  ├── Single-label multi-class? → Softmax + Cross-Entropy
  └── Multi-label? → Sigmoid + BCE (per label)

Is it regression?
  ├── Unbounded? → Linear + MSE/MAE
  ├── Positive only? → ReLU/Softplus + MSE
  └── Robust to outliers? → Linear + Huber/MAE

Is it probabilistic/generative?
  ├── Distribution matching? → Softmax + KL
  ├── GAN? → Various adversarial losses
  └── VAE? → Reconstruction + KL

Is it structured/sequence?
  ├── Language model? → Softmax + Token CE
  ├── Speech? → CTC Loss
  └── Similarity learning? → Linear + Triplet/Contrastive
```

## PART 5: Modern Defaults (2024-2026)

| Layer Type | Recommended Activation | Notes |
|------------|------------------------|-------|
| Hidden layers (CNN) | **SiLU / GELU / Mish** | Outperforms ReLU in most cases |
| Hidden layers (Transformer) | **GELU** | Standard in BERT, GPT, ViT |
| Hidden layers (deep CNN) | **ReLU / Leaky ReLU** | Still common, well-tested |
| Output (binary) | **Sigmoid** | Unchanged |
| Output (multi-class) | **Softmax** | Unchanged |
| Output (regression) | **Linear** | Unchanged |

| Scenario | Recommended Loss | Notes |
|----------|-----------------|-------|
| Imbalanced classification | **Focal Loss** | Down-weights easy examples |
| Medical segmentation | **Dice + CE** | Combines overlap and pixel accuracy |
| Self-supervised learning | **InfoNCE** | Standard for contrastive methods |
| Knowledge distillation | **KL Divergence** | Match teacher distribution |
| Training stability | **Label Smoothing + CE** | Prevents overconfidence |
